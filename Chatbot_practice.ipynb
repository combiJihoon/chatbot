```python
import openpyxl
import csv
```

```python
# 한 건의 대화에 대한 정보를 담는 객체입니다.
class Conversation :
  # 질문, 응답 두 변수로 구성됩니다.
  def __init__(self, contentName, contentType, question, answer) :
    # 생성자 만들기
    self. contentName = contentName
    self. contentType = contentType
    self. question = question
    self. answer = answer

  # 출력 정보 만들기
  def __str__(self) :
    return "질문: " + self.question + "\n답변: " + self.answer + "\n"
    
wb = openpyxl.load_workbook("/content/drive/MyDrive/Coding/Project/3. RecommendingBot/Input.xlsx")

ws = wb.active

conversations = []

# 시트 내에 존재하는 모든 데이터를 객체로 담습니다.
for r in ws.rows :
  c = Conversation(r[0].value, r[1].value, r[2].value, r[3].value)
  # 해당 객체를 담는다.
  conversations.append(c)

wb.close()

for c in conversations :
  print(str(c))
```
